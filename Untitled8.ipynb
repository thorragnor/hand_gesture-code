```python
import cv2
from cvzone.HandTrackingModule import HandDetector
import pywhatkit
import time

# Initialize webcam and hand detector
cap = cv2.VideoCapture(0)
detector = HandDetector(detectionCon=0.8, maxHands=1)

last_sent_time = 0
cooldown = 5  # seconds
recipient = "+916375******"  # Change to desired number

def send_message(message):
    global last_sent_time
    current_time = time.time()
    if current_time - last_sent_time > cooldown:
        try:
            pywhatkit.sendwhatmsg_instantly(
                phone_no=recipient,
                message=message,
                tab_close=True
            )
            print("Message Sent:", message)
            last_sent_time = current_time
        except Exception as e:
            print("Error sending message:", e)

while True:
    success, img = cap.read()
    img = cv2.flip(img, 1)  # Mirror the image

    hands, img = detector.findHands(img)

    if hands:
        hand = hands[0]
        fingers = detector.fingersUp(hand)

        # Match gestures
        if fingers == [1, 1, 1, 1, 1]:
            cv2.putText(img, "Gesture: All fingers up - Welcome Message", (10, 50),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
            send_message("Welcome to Linux World! ðŸ‘‹")

        elif fingers == [1, 0, 0, 0, 1]:
            cv2.putText(img, "Gesture: Call Me ", (10, 90),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
            send_message("call me")

        elif fingers == [0, 1, 0, 0, 0]:
            cv2.putText(img, "Gesture: Emergency h", (10, 130),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 2)
            send_message("I am in Danger")

        elif fingers == [0, 1, 1, 1, 1]:
            cv2.putText(img, "Gesture: High Five ", (10, 170),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)
            send_message("Emergency")

        elif fingers == [0, 1, 1, 0, 0]:
            cv2.putText(img, "Gesture: Peace ", (10, 210),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (200, 0, 200), 2)
            send_message("help me ")

    # Display the webcam image
    cv2.imshow("WhatsApp Gesture Control", img)
    key = cv2.waitKey(1)
    if key == 27:  # ESC to exit
        break

cap.release()
cv2.destroyAllWindows

```
